```pyspark
flights_tiny_df = flights_tiny_df \
    .withColumn("delayed_time2", flights_tiny_df["scheduled_depature_time"] - flights_tiny_df["actual_departure_time"])
```

```pyspark
flights_tiny_df = flights_tiny_df \
    .withColumn(
        "scheduled_departure_timestamp", 
        F.to_timestamp(flights_tiny_df["scheduled_depature_time"], "HH:mm:ss")
    )

flights_tiny_df = flights_tiny_df \
    .withColumn(
        "actual_departure_timestamp", 
        F.to_timestamp(flights_tiny_df["actual_departure_time"], "HH:mm:ss")
    )
```

```pyspark
from pyspark.sql.window import Window
```

```pyspark
country="United States"

flights_col = [
    "tail_number"
]
aircraft_col = [ 
    "manufacturer",
    "model"
]
airlines_col = [
    "carrier_code",
    "name"
]

result_col = aircraft_col + airlines_col + flights_col + ["sum(count)"]

airlines_all_counts = airlines_df.filter(airlines_df["country"] == country) \
    .join(flights_tiny_df, airlines_df.carrier_code == flights_tiny_df.carrier_code) \
    .drop(flights_tiny_df.carrier_code) \
    .groupBy("name", "carrier_code", "tail_number") \
    .count() \
    .join(aircraft_df, 
        flights_tiny_df.tail_number == aircraft_df.tailnum,
        how="left"
    ) \
    .filter(F.col("manufacturer").isNotNull()) \
    .filter(F.col("model").isNotNull()) \
    .groupBy("name", "manufacturer", "model") \
    .sum("count").cache()
airlines_all_counts.show()

# flights_tiny_df.select(flights_col)
#     .groupBy("tail_number") \
#     .count() \
#     .join(aircraft_df, aircraft_df.tailnum == flights_tiny_df.tail_number) \
#     .select()
#     .sort(["manufacturer", "model"]) \
#     .drop(aircraft_df.tailnum) \
#     .show()
```

```pyspark
window = Window.partitionBy(airlines_all_counts["name"]).orderBy(airlines_all_counts["sum(count)"])
```

```pyspark
airlines_all_counts.select("*", F.rank().over(window).alias("rank")) \
    .filter(F.col("rank") <=5) \
    .sort("name", ascending=True) \
    .show()
```
